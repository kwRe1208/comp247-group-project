```python
#explore each columns of the dataset
def analyze_columns(df, columns, plot_columns=None):
    """
    Analyzes the given columns of the DataFrame by showing their value counts,
    the number of null values, and optionally plots the value counts for specific columns.

    Parameters:
    - df: The DataFrame to analyze.
    - columns: A list of column names to analyze.
    - plot_columns: Optional. A list of column names for which to plot value counts.
    """
    if plot_columns is None:
        plot_columns = []
    
    for col in columns:
        print(f"Analysis for column: {col}")
        print("Value Counts:")
        print(df[col].value_counts())
        print("Number of null values:", df[col].isnull().sum())
        print("\n")
        if col in plot_columns:
            df[col].value_counts().plot(kind='bar')
            plt.title(f"Value Counts for {col}")
            plt.ylabel('Counts')
            plt.xlabel(col)
            plt.show()
```

```python
#use the analyze_columns function
columns_to_analyze = ['ACCLASS']
#plot_columns = ['VEHTYPE']  # Columns to plot
analyze_columns(df, columns_to_analyze)
```

```python
# find features with maximum missing values
threshold = df.shape[0] * 0.8
columns_with_nulls = df.columns[df.isnull().sum() > threshold]
columns_with_nulls
```

```python
# List of columns identified with more than 80% missing values you want to drop - based on the result above
columns_to_remove = ['OFFSET', 'FATAL_NO', 'PEDTYPE', 'PEDACT', 'PEDCOND', 
                     'CYCLISTYPE', 'CYCACT', 'CYCCOND']

# Drop these specific columns from the dataframe
df = df.drop(columns=columns_to_remove)

df.columns
```

```python
# drop irrelevant columns
# Find columns where the number of unique values is equal to the number of rows
unique_id_columns = [col for col in df.columns if df[col].nunique() == df.shape[0]]
df = df.drop(columns=unique_id_columns, axis=1)
df.shape
unique_id_columns
```

```python
# investigating target column - ACCLASS
# Change the property damage (target column) - to non-fatal
# Assuming data['HOOD_140'] contains the data you want to plot
accident_class_counts = df['ACCLASS'].value_counts()
# Plotting a bar plot of counts - for visualization
plt.figure(figsize=(5,5))
accident_class_counts.plot(kind='bar', color='skyblue')
plt.xlabel('ACCLASS')
plt.ylabel('Count')
plt.title('Distribution of ACCLASS')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# print unique values from ACCLASS
# Get unique values from the column
acclass_unique_values = df['ACCLASS'].unique()

# Print the unique values
print(f"\nUnique values of ACCLASS Column:")
print(acclass_unique_values)

# Replace 'Property Damage Only' with 'Non-Fatal Injury' in the 'ACCLASS' column
df['ACCLASS'] = df['ACCLASS'].replace('Property Damage Only', 'Non-Fatal Injury')

# Now check the unique values again to confirm the change
acclass_unique_values_after = df['ACCLASS'].unique()

# Print the unique values after the change
print(f"\nUnique values of ACCLASS Column after change:")
print(acclass_unique_values_after)
```

```python
# figuring out categorical columns
# Identify categorical columns by inspecting data types and unique value counts
categorical_columns = []

for column in df.columns:
    if df[column].dtype == 'object' or df[column].dtype.name == 'category' or df[column].nunique() / df.shape[0] < 0.05:
        categorical_columns.append(column)

print("\nPotential categorical columns:")
print(categorical_columns)
```

```python
# Define the categorical columns from potential categorical columns
categorical_columns = ['ROAD_CLASS', 'DISTRICT', 'VISIBILITY', 'RDSFCOND', 
                       'LOCCOORD', 'IMPACTYPE', 'TRAFFCTL', 'INVTYPE']

# Initialize an empty list to store row indices to drop
rows_to_drop = []

# Loop through each categorical column
for column in categorical_columns:
    # Find indices of rows where the column has a missing value
    missing_indices = df[df[column].isnull()].index
    # Add these indices to the list
    rows_to_drop.extend(missing_indices)

# Remove duplicates from the list of rows to drop
rows_to_drop = list(set(rows_to_drop))

# Drop the rows from the DataFrame
df = df.drop(index=rows_to_drop)
df.shape
```

```python
# Initialize an empty list to hold columns that contain "Yes" among other values
columns_with_yes = []

# Iterate over each column in the DataFrame
for column in df.columns:
    # Check if "Yes" is one of the unique values in the column
    if 'Yes' in df[column].unique().tolist():
        # If so, add the column name to the list
        columns_with_yes.append(column)

# Print the list of columns that contain "Yes" among other values
print("Columns that contain 'Yes' among other values:")
print(columns_with_yes)
```

```python
# List of columns that contain 'Yes' among other values and where you want to convert nulls to 'No'
columns_to_fill = ['PEDESTRIAN', 'CYCLIST', 'AUTOMOBILE', 'MOTORCYCLE', 'TRUCK', 
                   'TRSN_CITY_VEH', 'EMERG_VEH', 'PASSENGER', 'SPEEDING', 
                   'AG_DRIV', 'REDLIGHT', 'ALCOHOL', 'DISABILITY']

# Iterate over each column in the list and replace nulls with 'No'
for column in columns_to_fill:
    df[column] = df[column].fillna('No')

# Verify the operation by displaying the number of nulls in these columns after replacement
print(df[columns_to_fill].isnull().sum())
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])

# Extract month and date into new columns
df['Month'] = df['DATE'].dt.month
df['Day'] = df['DATE'].dt.day
df['Time'] = df['DATE'].dt.time
```

```python
accidents_per_year = df['YEAR'].value_counts().sort_index()

# Plot the graph
plt.figure(figsize=(8,4))
accidents_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
accidents_per_month = df['Month'].value_counts().sort_index()

# Plot the graph
plt.figure(figsize=(8, 4))
accidents_per_month.plot(kind='bar', color='skyblue')
plt.title('Number of Accidents by Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
accidents_per_month = df['Day'].value_counts().sort_index()

# Plot the graph
plt.figure(figsize=(8, 4))
accidents_per_month.plot(kind='bar', color='skyblue')
plt.title('Number of Accidents by Day')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extract day of the week into a new column (Monday=0, Sunday=6)
df['DayOfWeek'] = df['DATE'].dt.dayofweek

# Map the day of the week from numbers to names for readability
days = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 
        4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
df['DayOfWeekName'] = df['DayOfWeek'].map(days)

# Count accidents per day of the week
accidents_per_day_of_week = df['DayOfWeekName'].value_counts().reindex(list(days.values()))

# Plot the graph
plt.figure(figsize=(8, 4))
accidents_per_day_of_week.plot(kind='bar', color='skyblue')
plt.title('Number of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
